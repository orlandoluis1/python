```python
import numpy as np
import matplotlib.pyplot as plt

def potencial_eletrico(q, r0, x, y):
    """Calcula o potencial elétrico em um ponto (x, y) devido a uma carga q em r0."""
    k = 8.9875517873681764e9 # Constante de Coulomb
    r = np.sqrt((x - r0[0])**2 + (y - r0[1])**2)
    with np.errstate(divide='ignore', invalid='ignore'):
        V = k * q / r
        V[r == 0] = np.inf # Evita divisão por zero
    return V

# Configuração do espaço
x = np.linspace(-1, 1, 30)
y = np.linspace(-1, 1, 30)
X, Y = np.meshgrid(x, y)

# Configurações das cargas
cargas = [(-3e-9, (-0.5, 0)), (3e-9, (0.5, 0))]

# Calcula o potencial resultante
V = np.zeros_like(X)
for q, r0 in cargas:
    V += potencial_eletrico(q, r0, X, Y)

# Configuraçao das linhas de contorno das curvas equipotenciais
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, V, levels=50, cmap='viridis')
plt.colorbar(contour) # Barra de cores das linhas
plt.title('Curvas Equipotenciais') #Titulo
plt.xlabel('x') #eixo X
plt.ylabel('y') #eixo Y
plt.grid(True)
plt.axis('scaled') # Definir a proporção
plt.show() # Mostrar o gráfico das curvas

```
